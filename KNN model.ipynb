```python
import numpy as np
import pandas as pd

import scipy
import matplotlib.pyplot as plt  
from pylab import rcParams
import sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn import neighbors
from sklearn import preprocessing
import seaborn as sns
from sklearn.model_selection import train_test_split

from sklearn import metrics

df = pd.read_csv('D:\\IBM Project\\Weather Prediction\\Test Data\\Testclean2.csv')
```

```python
dummies = pd.get_dummies(df['_conds'])
dummies
```

```python
dummies2 = pd.get_dummies(df['_wdire'])
dummies2
```

```python
merged = pd.concat([df,dummies,dummies2], axis='columns')
merged
```

```python
final = merged.drop(['_conds','_wdire','Thunderstorms with Hail','SSW'],axis='columns')
final
```

```python
X = final.drop('datetime_utc',axis='columns')
X
```

```python
Y = final.datetime_utc
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
```

```python
print('Accuracy of KNN n-100, on the training set: {:.3f}'.format(knn.score(X_train, Y_train)))
print('Accuracy of KNN n-100, on the test set: {:.3f}'.format(knn.score(X_test, Y_test)))
```
